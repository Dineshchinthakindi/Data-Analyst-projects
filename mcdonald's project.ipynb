```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(data['Calories'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Distribution of Calories')
plt.show()
```
